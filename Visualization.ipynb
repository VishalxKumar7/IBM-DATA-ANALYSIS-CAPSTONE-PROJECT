### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
## Write your code here
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
df = df[df['ConvertedCompYearly'] < 1000000]
df = df.dropna(subset=['ConvertedCompYearly'])

plt.figure(figsize=(10,6))
plt.hist(df['ConvertedCompYearly'], bins=50, edgecolor='black')
plt.title('Distribution of Yearly Compensation')
plt.xlabel('Total Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
## Write your code here
df = df.dropna(subset=['Age'])
mapping_age_group = {
    "Under 18 years old": 18,
    "18-24 years old": (18+24)/2,
    "25-34 years old": (25+34)/2,
    "35-44 years old": (35+44)/2,
    "45-54 years old": (45+54)/2,
    "55-64 years old": (55+64)/2,
    "65 years or older": 70
}
df['Age_numeric'] = df['Age'].map(mapping_age_group)

plt.figure(figsize=(6, 5))
plt.boxplot(x=df['Age_numeric'], patch_artist=True)
plt.title('Box Plot of Age Distribution')
plt.ylabel('Age')
plt.show()
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
## Write your code here

plt.figure(figsize=(8,5))
plt.scatter(x=df['Age_numeric'], y=df['ConvertedCompYearly'])
plt.title('Relationship Between Age and Compensation')
plt.xlabel('Age')
plt.ylabel('Yearly Compensation')
plt.grid()
plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
## Write your code here
df= df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'Age_numeric'])

plt.figure(figsize=(12,5))
plt.scatter(x=df['JobSatPoints_6'], y=df['ConvertedCompYearly'], s=df['Age_numeric']*5)
plt.title('Bubble Plot of Compensation vs Job Satisfaction (Bubble Size: Age)')
plt.xlabel('Job SatisFaction')
plt.ylabel('Yearly Compensation')
plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
## Write your code here
df['MainBranch'] = df['MainBranch'].replace('I am not primarily a developer, but I write code sometimes as part of my work/studies', "I code occasionally for work/studies.")
```

```python
df = df.dropna(subset=['MainBranch'])
mainbranch_count = df['MainBranch'].value_counts()

plt.figure(figsize=(8,4))
plt.barh(mainbranch_count.index, mainbranch_count.values, height=0.4)
plt.title('Distribution of Respondents by MainBranch')
plt.xlabel('Number of Respondents')
plt.ylabel('Primary Role (MainBranch)')
plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
## Write your code here
df = df.dropna(subset=['LanguageWantToWorkWith'])
df_explode = df['LanguageWantToWorkWith'].str.split(';').explode()
lang_counts = df_explode.value_counts().sort_values(ascending=False).reset_index().head(5)
lang_counts.columns = ['Language', 'Count']

plt.figure(figsize=(8,5))
plt.bar(lang_counts['Language'], lang_counts['Count'], width=0.6)
plt.title('Top 5 Programming Language')
plt.xlabel('Language')
plt.ylabel('Number of Respondents')
plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
## Write your code here
df = df.dropna(subset=['JobSatPoints_6', 'JobSatPoints_7', 'Age'])
df_group = df.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

df_group.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title('Stacked Bar Chart of Job Satisfaction Scores by Age')
plt.xlabel('Age')
plt.ylabel('Median Job Satisfaction')
plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
## Write your code here
df = df.dropna(subset=['DatabaseHaveWorkedWith'])
df_explode = df['DatabaseHaveWorkedWith'].str.split(';').explode()
db_count = df_explode.value_counts().sort_values(ascending=False).head(5).reset_index()
db_count.columns = ['Database', 'Count']

plt.figure(figsize=(10,5))
plt.bar(db_count['Database'], db_count['Count'], width=0.5)
plt.title('Top 5 Most Used DataBase')
plt.xlabel('DataBase')
plt.ylabel('Number of Respondents')
plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
## Write your code here
df_group = df.groupby('Age')['ConvertedCompYearly'].median()

plt.figure(figsize=(8,5))
plt.bar(df_group.index, df_group.values)
plt.xticks(rotation=90)
plt.title('Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Meidan Respondents (Yearly Compensation)')
plt.show()
```

##### 2. Bar Chart of Respondent Count by Country


Show the distribution of respondents by country to see which regions are most represented.



```python
## Write your code here
df = df.dropna(subset=['Country'])

df_group = df['Country'].value_counts().sort_values(ascending=False).reset_index()
df_group.columns = ['Country', 'Count']
df_group['Country'] = df_group['Country'].replace({'United States of America':'USA', 'United Kingdom of Great Britain and Northern Ireland':'UK'})
df_group = df_group.head(10)

plt.figure(figsize=(12,5))
plt.bar(df_group['Country'], df_group['Count'])
plt.title('Top 10 Country by Number of Respondents')
plt.xlabel('Country')
plt.ylabel('Number of Respondents')
plt.show()
```
